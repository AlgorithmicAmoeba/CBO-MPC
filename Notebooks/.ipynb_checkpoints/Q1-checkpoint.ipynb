```python
import numpy
import matplotlib.pyplot as plt
```

```python
def G(s):
    G11 = 12.8 * numpy.exp(-s) / (16.7*s + 1)
    G12 = -18.9 * numpy.exp(-3*s) / (21*s + 1)
    G21 = 6.6 * numpy.exp(-7*s) / (10.9*s + 1)
    G22 = -19.4 * numpy.exp(-3*s) / (14.4*s + 1)
    return numpy.matrix([[G11, G12], [G21, G22]])

def Gnd(s):
    G11 = 12.8 / (16.7*s + 1)
    G12 = -18.9 / (21*s + 1)
    G21 = 6.6 / (10.9*s + 1)
    G22 = -19.4 / (14.4*s + 1)
    return numpy.matrix([[G11, G12], [G21, G22]])
```

### (a)

```python
# %matplotlib inline
ws = numpy.logspace(-4, 2 , 1000)
nyqust = [numpy.linalg.det(numpy.eye(2) + G(1j*w)) for w in ws]

plt.gcf().set_size_inches(15, 10)

plt.subplot(2, 3, 1)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')

plt.subplot(2, 3, 2)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')
plt.xlim(-20, 30)
plt.ylim(-20, 20)

plt.subplot(2, 3, 3)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.subplot(2, 3, 4)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.subplot(2, 3, 5)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')
plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.subplot(2, 3, 6)
plt.plot(numpy.real(nyqust), numpy.imag(nyqust))
plt.plot(numpy.real(nyqust), -numpy.imag(nyqust))
plt.plot(0, 0, 'rx')
plt.xlim(-0.3, 0.3)
plt.ylim(-0.3, 0.3)

plt.savefig("nyquist.pdf")
plt.show()
```

```python
nyqust_nd = [numpy.linalg.det(numpy.eye(2) + Gnd(1j*w)) for w in ws]

plt.gcf().set_size_inches(15, 5)

plt.subplot(1, 3, 1)
plt.plot(numpy.real(nyqust_nd), numpy.imag(nyqust_nd))
plt.plot(numpy.real(nyqust_nd), -numpy.imag(nyqust_nd))
plt.plot(0, 0, 'rx')

plt.subplot(1, 3, 2)
plt.plot(numpy.real(nyqust_nd), numpy.imag(nyqust_nd))
plt.plot(numpy.real(nyqust_nd), -numpy.imag(nyqust_nd))
plt.plot(0, 0, 'rx')
plt.xlim(-20, 30)
plt.ylim(-20, 20)

plt.subplot(1, 3, 3)
plt.plot(numpy.real(nyqust_nd), numpy.imag(nyqust_nd))
plt.plot(numpy.real(nyqust_nd), -numpy.imag(nyqust_nd))
plt.plot(0, 0, 'rx')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.savefig("nyquist_nd.pdf")
plt.show()
```

### (b)

```python
ws = numpy.logspace(-2, 2 , 1000)
char_loci = [sorted(numpy.linalg.eigvals(G(1j*w))) for w in ws]


plt.plot(numpy.real(char_loci), numpy.imag(char_loci), '.')
plt.gcf().gca().add_artist(plt.Circle((0, 0), 1, fill=False, color='g'))

plt.gcf().set_size_inches(5, 5)
plt.xlim(-15, 10)
plt.ylim(-15, 10)

plt.savefig("locus.pdf")
plt.show()
```

```python
char_loci_nd = [sorted(numpy.linalg.eigvals(Gnd(1j*w))) for w in ws]


plt.plot(numpy.real(char_loci_nd), numpy.imag(char_loci_nd), '.')
plt.gcf().gca().add_artist(plt.Circle((0, 0), 1, fill=False, color='g'))
plt.gcf().set_size_inches(5, 5)
plt.xlim(-15, 10)
plt.ylim(-15, 10)

plt.savefig("locus_nd.pdf")
plt.show()
```
